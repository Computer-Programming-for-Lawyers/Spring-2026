# Modules in Python



A module is a Python file (usually ending in `.py`) that stores reusable code like functions, variables, and classes.

You import a module when you want to use that code in another file or notebook, instead of rewriting it.

Using modules keeps code organized, easier to test, and easier to maintain.

## Common import styles

You will commonly see three patterns:

1. `import helper` # imports entire module
2. `from helper import validate` # imports just the `validate` function
3. `import helper as hp` # imports entire module with an alias

Your module should only contain function and class definitions, and avoid running code at the top level. This makes it easier to import without side effects.

## Your modules vs libraries
There are lots of useful libraries you can install with `pip` (e.g., `requests` for web requests, `pandas` for data analysis). You can also create your own modules for code you want to reuse across projects.

When you import a module, Python looks for it in the current directory first. If it doesn't find it there, it looks in the standard library and then in installed packages.

## Troubleshooting imports

If an import fails, check:
- The module file is in the same folder as your notebook/script
- The filename is spelled correctly (`helpers.py` vs `helper.py`)

When you edit a `.py` module during notebook work, remember:
- Restarting the kernel is the most reliable way to clear old imports
- You can also reload the module without a full restart

Reload pattern:

```python
import importlib
import my_module
importlib.reload(my_module)
```

Keep function files focused on definitions (`def`, constants, classes).
