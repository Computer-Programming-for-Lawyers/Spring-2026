## List slicing refresh

Hint: Use negative slices to return a string backwards: **[::-1]**. Follows model of **[start:stop:step]**.


```python
# List slicing refresh

cases = [
    "Smith v. Jones",
    "Brown v. Board",
    "Roe v. Wade",
    "Miranda v. Arizona",
    "Marbury v. Madison",
    "Obergefell v. Hodges"
]

# print the second case
print(cases[1])

# print the 2nd through 4th cases
print(cases[1:4])

# print the last 3 cases
print(cases[-3:])

# print the cases in reverse order
print(cases[::-1])
```

## Dictionaries Refresh

* You're building a case management system that tracks details about each open case.
* Our goal is to create a dictionary, access values, modify entries, and iterate through key-value pairs.


```python
# Create a dictionary for a single case
case_info = {
    "case_id": "2026-001",
    "case_name": "Smith v. Jones",
    "status": "pending",
    "attorney": "Sarah Chen",
    "filing_date": "2026-01-10"
}

print("Case Information:")
print(case_info)
```

```python
# How would you access the case name?
case_info["case_name"]
```

```python
# How would you update the status to "closed"?
case_info["status"] = "closed"
```

```python
case_info
```

```python
# How would you add a new key-value pair for "next_hearing_date" 
# with the value "2026-07-15"?

case_info["next_hearing_date"] = "2026-07-15"

case_info
```

```python
# How would you iterate over the dictionary to print out all the 
# keys and values? (Hint: use a for loop)

for key in case_info:
    print(f"{key}"): {case_info[key]}")
    # print(key+": "+case_info[key]) # without f string
```

```python
for value in case_info.values():
    print(value)
```
