```python
import pandas as pd # import the pandas library

# read the csv into a pandas dataframe
birds = pd.read_csv('audubon_birds.csv')

# display two random rows of the dataframe
birds.sample(2)
```

```python
# select just the "conservation_status" column and store it as a list
conservation_status = birds["conservation_status"]
print(conservation_status)
```

## Your Task
Using a dictionary, iterate over the list "conservation_status" and create a count of how many birds are in each status.


```python
bird_conditions = {}

# for each item in the conservation_status list
for status in conservation_status:

    # check if that status is already a "key" in bird_conditions
    if status in bird_conditions:
        # if it is, update the value to itself + 1
        bird_conditions[status] += 1

    else: # if its not a key yet, add it to the dictionary and start the counter at 1
        bird_conditions[status] = 1

bird_conditions
```

Now,  print the names of birds that are "endangered" using list slicing.

```python
# starter lists
bird_names = birds["bird_name"]
conservation_status = birds["conservation_status"]

# loop through index position of bird_names
for i in range(len(bird_names)):
    #check whether the conservation status at the position under consideration is endangered
    if conservation_status[i] == "Endangered":
        #print bird name at that position if so
        print(bird_names[i])
```
