We haven't learned how to work with csv's yet, but the most common library used for working with tabular data (data organized in rows and columns, like a spreadsheet) is **Pandas**. As a preview to Pandas, you can see how the csv file looks as a "Pandas dataframe"  below. Today, we'll work with just one column of it in the form of a list.

(Note: You can't use Pandas in PS5.)

```python
import pandas as pd # import the pandas library

# read the csv into a pandas dataframe
birds = pd.read_csv('audubon_birds.csv')

# display two random rows of the dataframe
birds.sample(2)
```

```python
# select just the "conservation_status" column and store it as a list
conservation_status = birds["conservation_status"]
print(conservation_status)
```
