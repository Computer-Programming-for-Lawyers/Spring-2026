## word checker

A word is a word or prhase that reads the same backwards as it does forward, for example:
* "Never odd or even"
* "Level" 
* "Dad"
* "Civic"

Write a program that iterates over the txt file "wiki.txt" and returns all palindromes. The text file is the text from the Wikipedia page, [Palindrome](https://en.wikipedia.org/wiki/Palindrome).

*Hint: Use negative slices to return a string backwards: [::-1] (follows model of [start:stop:step])*

```python
wiki = open('wiki.txt').read().split()

# return any words that are palindromes
# words must be longer than one character, and not be numbers

# print(wiki)

# for each item in the list
for word in wiki:
    if len(word) > 1: # check if it's greater than one
        if word == word[::-1]: #check if it's the same backward
            if word.isalpha(): #check if it is just letters
                print(word) #print words that meet these conditions


# nested if statements one way to solve this -- could also do more complex boolean logic -- ie, if (len(word)>1) and (word == word[::-1])
```

```python
# more advanced: also look for four word palindromes, like "step on no pets"
# convert to lowercase, and don't include periods/spaces in your final output 

# create an empty list to add cleaned words to
clean_wiki = []

# loop through each item (word) in the wiki list
for word in wiki:
    clean_word = word.replace(",","").replace(".","").replace("!","").replace("|","") #replace some basic punctuation
    clean_word = clean_word.lower() #make the word lowercase
    clean_wiki += [clean_word] #add the item to the list


# loop through clean_wiki by INDEX POSITION (aka, starting at the 0th element and going until the last element)
for i in range(len(clean_wiki)):
    # print(i) # test what "i" is

    combined = clean_wiki[i] + clean_wiki[i+1] + clean_wiki[i+2] + clean_wiki[i+3] # we get an error here because we're going beyond the final list element
                                                                                   # you can fix this using the modulus operator we learned last week, we aren't doing that now for the sake of time

    if combined == combined[::-1]:
        print(combined) 
```
