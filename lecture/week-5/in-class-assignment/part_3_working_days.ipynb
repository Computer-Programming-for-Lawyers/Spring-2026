### **Task: Count how many days fall into each category**  

* Iterate through `adjournment_strings` using a **for loop**.

* Use **if & elif statements** to determine whether the session was **pro forma, not in session, or other**. Count it as pro forma if the string includes "pro forma", then as not in session if the string includes "not in session", then as other for all other categories.

* Hint: Convert the string to lowercase to make it easier to check for the presence of certain words.

```python
# starter dictionary
house = {"pro forma": 0, "not in session": 0, "other": 0}
senate = {"pro forma": 0, "not in session": 0, "other": 0}

```

```python
#create list of lowercase
adjournment_strings_lower = []

for item in adjournment_strings:
    adjournment_strings_lower += item.lower()
```

```python
for i in range(len(adjournment_strings_lower)): #don't need to use range/len here -- could also do for string in adjournmnet_strings...

    if "House" in adjournment_strings_lower[i]:

        if "not in session" in adjournment_strings_lower[i]: #check not in session first to address issues where pro forma is tomorrow
            house["not in session"] += 1

        elif "pro forma" in adjournment_strings_lower[i]:
            house["pro forma"] += 1

        else:
            house["other"] += 1

    else: # could also do elif "Senate" in ...

        if "not in session" in adjournment_strings_lower[i]:
            senate["not in session"] += 1

        elif "pro forma" in adjournment_strings_lower[i]:
            senate["pro forma"] += 1

        else:
            senate["other"] += 1
```
