```python


def extract_reporting_info(job_description):
    response = client.messages.create(
        model="claude-opus-4-6",
        max_tokens=1024,
        messages=[
            {
                "role": "user",
                "content": (
                    "Extract the following from this job description:\n"
                    "1) who the position reports to\n"
                    "2) the most important responsibility\n\n"
                    "Return only JSON matching the schema.\n\n"
                    "Job description:\n" + job_description
                ),
            }
        ],
        output_config={
            "format": {
                "type": "json_schema",
                "schema": {
                    "type": "object",
                    "properties": {
                        "reports_to": {"type": "string"},
                        "top_responsibility": {"type": "string"},
                    },
                    "required": ["reports_to", "top_responsibility"],
                    "additionalProperties": False,
                },
            }
        },
    )
    text = response.content[0].text
    try:
        return json.loads(text)
    except json.JSONDecodeError:
        raise ValueError(f"Unable to parse structured output: {text}")

# Example usage:
# result = extract_reporting_info(long_job_description)
# print(result)
```

```python
def safe_extract_reporting_info(description):
    if not isinstance(description, str):
        return {"reports_to": "", "top_responsibility": ""}
    try:
        return extract_reporting_info(description)
    except Exception as e:
        return {"reports_to": "", "top_responsibility": ""}

extracted = df["job_description"].apply(safe_extract_reporting_info)
df["reports_to"] = extracted.map(lambda item: item.get("reports_to", ""))
df["top_responsibility"] = extracted.map(lambda item: item.get("top_responsibility", ""))

df.head()
```
