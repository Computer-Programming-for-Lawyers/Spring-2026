# Week 4 Lecture Notebook: House Pattern

The "house pattern" generally refers to a way of transforming elements in a list into another list by applying a specific operation to each element. The pattern involves:

1. **Input list**: Start with an iterable (such as a list or range).
2. **Processing**: Loop through each element and apply a defined transformation (e.g., squaring, doubling, or any other operation).
3. **Output list**: Store the results of each transformation in a new list.

This pattern is typically achieved using a `for` loop. The key idea is that you process every element in the input iterable and generate a new list based on the applied operation.


```python
# given a list of numbers from 0 to 10, create a new list with that number times 1,000

list_of_numbers = list(range(0,11))

transformed_list = []

for numb in list_of_numbers:
    transformed_list += [numb*1000]

print(transformed_list)
```

```python
# Given a list of words brandeis.txt 
# write a house pattern filter that will produce a list containing only the words that have upper-case letters in them.

words = open('brandeis.txt').read().split() 

# hint: use .lower method to check if word is equal to lowercase version
words
```

```python
uppercase_values = []

for word in words:
    if not word == word.lower():
        uppercase_values += [word]

uppercase_values
```
