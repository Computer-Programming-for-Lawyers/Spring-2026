# Intro to Lists

Lists let us store multiple values in one variable and access them by position.
We’ll practice creating lists, checking their length, and pulling out specific items.
A list’s type depends on its content (e.g., a "list of integers" or "list of strings").

```python
# how to create lists

numbers = [3, 5, 7, 9]
names = ["Ada","Grace","Linus"]

list("cat")
```

```python
# lists can store many kinds of data (keep types consistent)

["cat",5,5.66]
```

## Range()
Range is a helpful list creation function we'll use a lot in this course. Range creates a full list of ordered integers, based on your specifications.


```python
# examples of range output

list(range(5))
```

```python
# create index positions for a list

items = ["pencil","pen","eraser"]

list(range(len(items)))
```

```python
# range by itself returns a generator, not a list

range(5)
```

```python
# range(start, stop): stop is not included

list(range(100,111))
```

```python
# range(start, stop, step): count by step size

list(range(0,20,2))
```
